```python
#gli algoritmi di machine learning non possono lavorare con "object" quindi riscriviamo la colonna date come datetime
matches["date"] = pd.to_datetime(matches["date"])
matches.dtypes
```

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes #.cat.codes convert into integers da category type
matches
#in questo modo abbiamo 0 se è fuori casa e 1 se è a casa per rendere numerico il venue
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes #ogni avversario ha un proprio codice
matches
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int") #convertiamo le ore in un numero togliendo i minuti sostituisce i minuti con lo spazio vuoto
matches["day_code"] = matches["date"].dt.dayofweek #assegno un numero al giorno della settimana
matches
```

```python
matches["target"] = (matches["result"] == "W").astype("int") #impostiamo il target quindi se la squadra vince -->1 se perde o pareggia -->0
matches
```

```python
#train machine learning model importiamo RandomForest da skitlearn
from sklearn.ensemble import RandomForestClassifier
#inizializzo la classe Randomforest con n_estimators è il numero di numero alberi di decisione che vogliamo addestrare
#min_samples_split numero di nodi che voglio sulla foglia prima dello split
#random_state = 1 stesso risultato se lo runniamo piu volte
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
#training set prima del 2022 test dopo 2022
train = matches[matches["date"] < '2022-01-01']
test = matches[matches["date"] > '2022-01-01']
#lista della predictors column
predictors = ["venue_code", "opp_code", "hour", "day_code"]
#fit
rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(test["target"], preds)
```

```python
acc
```

```python
#creiamo questo data frame per capire quando l'accuracy è alta o bassa
combined = pd.DataFrame(dict(actual = test["target"], prediction = preds))
#con pandas creiamo una tabella per vedere cosa succede quando viene predetta una partita se è corretta o meno
pd.crosstab(index=combined["actual"], columns=combined["prediction"])
```

```python
#non sono soddisfacienti i risultati quindi usiamo la precision
from sklearn.metrics import precision_score

precision_score(test["target"], preds)
```

```python
#cerchiamo di migliorare la precision
grouped_matches = matches.groupby("team") #raggruppo le partite per team
```

```python
#funzione che ordina per data le partite per poi fare una media nelle giornate precedenti a quelli chevoglio predire
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean() #prende una finestra pari a 3 e fa la media.
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols) #rimuove le righe relative ai missing values
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
group = grouped_matches.get_group("Manchester City").sort_values("date")
rolling_averages(group, cols, new_cols)
```

```python
#applichiamo a tutti i match una funzione per ogni team
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling
```

```python
#levo il primo livello di indice in modo sia piu facile accederci
matches_rolling = matches_rolling.droplevel('team')
matches_rolling
```

```python
#nuova numerazione unica dell'indice da 0 a 1316
matches_rolling.index = range(matches_rolling.shape[0])
matches_rolling
```

```python
#facciamo quello che abbiamo fatto prima con questa funzione
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    precision = precision_score(test["target"], preds)
    return combined, precision

combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```

```python
precision  #migliorata
```

```python

combined = combined.merge(matches_rolling[["date", "team", "opponent", "result"]], left_index=True, right_index=True)
```

```python
combined
```

```python
#combiniamo le partite come squadra casa e traferta della stessa partita insieme
#creo un dizionario
class MissingDict(dict):
    __missing__ = lambda self, key: key

map_values = {"Brighton and Hove Albion": "Brighton", "Manchester United": "Manchester Utd", "Newcastle United": "Newcastle Utd", "Tottenham Hotspur": "Tottenham", "West Ham United": "West Ham", "Wolverhampton Wanderers": "Wolves"}
mapping = MissingDict(**map_values)
combined["new_team"] = combined["team"].map(mapping)
```

```python
#merge per unica partita casa/traferta
merged = combined.merge(combined, left_on=["date", "new_team"], right_on=["date", "opponent"])
merged
```

```python
#guardiamo le righe dove uno perde e uno vince e guuardiamo il calore attuale di x
merged[(merged["predicted_x"] == 1) & (merged["predicted_y"] ==0)]["actual_x"].value_counts()
```
